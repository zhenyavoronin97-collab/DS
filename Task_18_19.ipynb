```python
# Сложность O(1)

def print_students_info(students):
  for student in students:
    print(f"Имя: {student['name']}, Возраст: {student['age']}, Курс: {student['course']}")
students = [
    {"name": "Евгений", "age": 27, "course": 2},
    {"name": "Андрей", "age": 26, "course": 1},
    {"name": "Павел", "age": 31, "course": 4}
]
print_students_info(students)



```
